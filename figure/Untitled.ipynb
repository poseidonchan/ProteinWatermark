```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Extended list of tokens representing amino acids
tokens = ['A', 'R', 'N']

# Assuming a uniform distribution, each amino acid has an equal probability
uniform_probability = 1 / len(tokens)
probabilities = [0.6, 0.1, 0.3]
```

```python
sns.set(style="white")

# Creating the plot with seaborn
fig, ax = plt.subplots(1, 1, figsize=(4, 4),  sharey=True)
sns.barplot(x=tokens, y=probabilities, color='#51A9D6', ax=ax)


plt.xlabel('Tokens (Amino Acids)', fontsize=14)
plt.ylabel('Probability', fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig("original_dist.pdf", format='pdf')
plt.show()
```

```python
sns.set(style="white")

# Creating the plot with seaborn
fig, ax = plt.subplots(1, 1, figsize=(4, 4),  sharey=True)
sns.barplot(x=tokens, y=[0.7, 0.05, 0.25], color='#FF653A', ax=ax)


plt.xlabel('Tokens (Amino Acids)', fontsize=14)
plt.ylabel('Probability', fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig("watermarked_dist1.pdf", format='pdf')
plt.show()
```
